```python
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()

# Build graph using TF operations(설계)

# y = x 인 데이터
x_train = [1, 2, 3]
y_train = [1, 2, 3]

# Variable 이라는 node
# tensorflow가 변경시키는 값이라고 생각하자
W = tf.Variable(tf.random_normal([1]), name='weight') # shape이 1인 임의값
b = tf.Variable(tf.random_normal([1]), name='bias')

# our hypothesis X * w + b
hypothesis = x_train * W + b
```

```python
t = [1., 2., 3., 4.]
print(tf.reduce_mean(t))
```

```python
cost = tf.reduce_mean(tf.square(hypothesis - y_train))
```

```python
# Minimize
optimizer = tf.train.GradientDescentOptimizer(learning_rate = 0.01)
train = optimizer.minimize(cost)
```

```python
# Session을 통해서 graph에 데이터를 넣는다
sess = tf.Session()
sess.run(tf.global_variables_initializer()) # variable을 사용하기 전에는 반드시 initialize를 해줘야 한다

for step in range(2001):
    sess.run(train)
    if step % 20 == 0:
        print(step, sess.run(cost), sess.run(W), sess.run(b))
```

```python
# placeholder을 이용해보자
# linear regression 공식을 사용하고, 값을 나중에 placeholder로 줄 수 있기 때문에 편하다
X = tf.placeholder(tf.float32)
Y = tf.placeholder(tf.float32)

for step in range(2001):
    cost_val, W_val, b_val, _ = sess.run([cost, W, b, train], feed_dict = {X: [1, 2, 3], Y: [1, 2, 3]})
    if step % 20 == 0:
        print(step, cost_val, W_val, b_val)
```
