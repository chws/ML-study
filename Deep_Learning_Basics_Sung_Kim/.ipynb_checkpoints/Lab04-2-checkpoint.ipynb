```python
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()
import numpy as np
tf.set_random_seed(777) # for reproducibility
```

```python
xy = np.loadtxt('data-01-test-score.csv', delimiter=",", dtype=np.float32) # 파일 읽어오기
x_data = xy[:, 0:-1] # n개의 행을 전부 다 가져오고, 그 행들에서 0부터 마지막 값 제외한 것까지 가져오기
y_data = xy[:, [-1]] # n개의 행을 전부 다 가져오는데, 마지막 값만 y로 치겠다
```

```python
# Make sure the shape and data are ok
print(x_data.shape, x_data, len(x_data))
print(y_data.shape, y_data)
```

```python
# placeholders for a tensor that will be always fed
X = tf.placeholder(tf.float32, shape=[None, 3])
Y = tf.placeholder(tf.float32, shape=[None, 1])
```

```python
W = tf.Variable(tf.random_normal([3, 1]), name='weight')
b = tf.Variable(tf.random_normal([1]), name='bias')
```

```python
# Hypothesis
hypothesis = tf.matmul(X, W) + b
```

```python
# Simplified cost/Loss function
cost = tf.reduce_mean(tf.square(hypothesis - Y))
```

```python
# Minimize
optimizer = tf.train.GradientDescentOptimizer(learning_rate=1e-5)
train = optimizer.minimize(cost)
```

```python
# Launch the graph in a session
sess = tf.Session()
# Initializes global variables in the graph
sess.run(tf.global_variables_initializer())
# Set up feed_dict variables inside the loop
for step in range(2001):
    cost_val, hy_val, _ = sess.run([cost, hypothesis, train], feed_dict={X: x_data, Y: y_data})
    if step % 10 == 0:
        print(step, "Cost: ", cost_val, "\nPrediction:\n", hy_val)
```

```python
# Ask my score 학습한 것을 바탕으로 예측해보기
print("Your score will be ", sess.run(hypothesis, feed_dict={X: [[100, 70, 101]]}))
print("Other scores will be ", sess.run(hypothesis, feed_dict={X: [[60, 70, 110], [90, 100, 80]]}))
```
