```python
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()
import numpy as np
```

```python
x_data = np.array([[0, 0], [0, 1], [1, 0], [1, 1]], dtype=np.float32)
y_data = np.array([[0], [1], [1], [0]], dtype=np.float32)
```

```python
X = tf.placeholder(tf.float32)
Y = tf.placeholder(tf.float32)
W = tf.Variable(tf.random_normal([2, 1]), name='weight') # in: 2, out: 1
b = tf.Variable(tf.random_normal([1]), name='bias') # out: 1
```

```python
# 일반 Logistic Regression의 경우
# Hypothesis using sigmoid: tf.div(1., 1. + tf.exp(tf.matmul(X, W)))
hypothesis = tf.sigmoid(tf.matmul(X, W) + b)

# cost/loss function
cost = -tf.reduce_mean(Y * tf.log(hypothesis) + (1 - Y) * tf.log(1 - hypothesis))
train = tf.train.GradientDescentOptimizer(learning_rate=0.1).minimize(cost)
```

```python
# Accuracy computation
# True if hypothesis>0.5 else False
predicted = tf.cast(hypothesis > 0.5, dtype=tf.float32)
accuracy = tf.reduce_mean(tf.cast(tf.equal(predicted, Y), dtype=tf.float32)) # Y 값과 같은지 아닌지 T, F로 결과를 낸다
```

```python
# Launch graph
with tf.Session() as sess:
    # Initialize Tensorflow variables
    sess.run(tf.global_variables_initializer())
    
    for step in range(10001):
        sess.run(train, feed_dict={X: x_data, Y: y_data})
        if step % 100 == 0:
            print(step, sess.run(cost, feed_dict={X: x_data, Y: y_data}), sess.run(W))
            
    # Accuracy report
    h, c, a = sess.run([hypothesis, predicted, accuracy], feed_dict={X: x_data, Y: y_data})
    print("\nHypothesis: ", h, "\nCorrect: ", c, "\nAccuracy: ", a) # Accuracy가 0.5 밖에 안된다!
```

```python
# first layer
W1 = tf.Variable(tf.random_normal([2, 2]), name='weight1') # input: 2, output: 2 -> layer의 output이 2라는 의미
b1 = tf.Variable(tf.random_normal([2]), name='bias1') # bias는 output과 같은 2
layer1 = tf.sigmoid(tf.matmul(X, W1) + b1) # 첫번째 레이어에는 X가 들어간다
```

```python
W2 = tf.Variable(tf.random_normal([2, 1]), name='weight2') # layer1의 output이 2이므로, 출력은 Y로, 1로 정해져 있으므로 out=1
b2 = tf.Variable(tf.random_normal([2]), name='bias2')
hypothesis = tf.sigmoid(tf.matmul(layer1, W2) + b2) # 첫번째 레이어를 두번째 레이어에 넣어준다
```

```python
# cost/loss function
cost = -tf.reduce_mean(Y * tf.log(hypothesis) + (1 - Y) * tf.log(1 - hypothesis))
train = tf.train.GradientDescentOptimizer(learning_rate=0.1).minimize(cost)
```

```python
# Accuracy computation
# True if hypothesis>0.5 else False
predicted = tf.cast(hypothesis > 0.5, dtype=tf.float32)
accuracy = tf.reduce_mean(tf.cast(tf.equal(predicted, Y), dtype=tf.float32)) # Y 값과 같은지 아닌지 T, F로 결과를 낸다
```

```python
# Launch graph
with tf.Session() as sess:
    # Initialize Tensorflow variables
    sess.run(tf.global_variables_initializer())
    
    for step in range(10001):
        sess.run(train, feed_dict={X: x_data, Y: y_data})
        if step % 100 == 0:
            print(step, sess.run(cost, feed_dict={X: x_data, Y: y_data}), sess.run([W1, W2]))
            
    # Accuracy report
    h, c, a = sess.run([hypothesis, predicted, accuracy], feed_dict={X: x_data, Y: y_data})
    print("\nHypothesis: ", h, "\nCorrect: ", c, "\nAccuracy: ", a) # layer을 2개로 NN을 이용했더니 Accuracy가 1.0으로 증가하였다!
```

```python
# Wide NN for XOR
# 첫번째 out을 10으로 주게 되면, 작은 값은 더 작아지고, 큰 값은 더 커졌다-> 더 정확해짐
# first layer
W1 = tf.Variable(tf.random_normal([2, 10]), name='weight1') # input: 2, output: 2 -> layer의 output이 2라는 의미
b1 = tf.Variable(tf.random_normal([10]), name='bias1') # bias는 output과 같은 2
layer1 = tf.sigmoid(tf.matmul(X, W1) + b1) # 첫번째 레이어에는 X가 들어간다

W2 = tf.Variable(tf.random_normal([10, 1]), name='weight2') # layer1의 output이 2이므로, 출력은 Y로, 1로 정해져 있으므로 out=1
b2 = tf.Variable(tf.random_normal([2]), name='bias2')
hypothesis = tf.sigmoid(tf.matmul(layer1, W2) + b2) # 첫번째 레이어를 두번째 레이어에 넣어준다

# cost/loss function
cost = -tf.reduce_mean(Y * tf.log(hypothesis) + (1 - Y) * tf.log(1 - hypothesis))
train = tf.train.GradientDescentOptimizer(learning_rate=0.1).minimize(cost)

# Accuracy computation
# True if hypothesis>0.5 else False
predicted = tf.cast(hypothesis > 0.5, dtype=tf.float32)
accuracy = tf.reduce_mean(tf.cast(tf.equal(predicted, Y), dtype=tf.float32)) # Y 값과 같은지 아닌지 T, F로 결과를 낸다

# Launch graph
with tf.Session() as sess:
    # Initialize Tensorflow variables
    sess.run(tf.global_variables_initializer())
    
    for step in range(10001):
        sess.run(train, feed_dict={X: x_data, Y: y_data})
        if step % 100 == 0:
            print(step, sess.run(cost, feed_dict={X: x_data, Y: y_data}), sess.run([W1, W2]))
            
    # Accuracy report
    h, c, a = sess.run([hypothesis, predicted, accuracy], feed_dict={X: x_data, Y: y_data})
    print("\nHypothesis: ", h, "\nCorrect: ", c, "\nAccuracy: ", a) # layer을 2개로 NN을 이용했더니 Accuracy가 1.0으로 증가하였다!
```
