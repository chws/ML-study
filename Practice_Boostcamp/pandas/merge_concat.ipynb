```python
import pandas as pd
```

```python
raw_data = {
        'subject_id': ['1', '2', '3', '4', '5', '7', '8', '9', '10', '11'],
        'test_score': [51, 15, 15, 61, 16, 14, 15, 1, 61, 16]}
df_a = pd.DataFrame(raw_data, columns = ['subject_id', 'test_score'])
df_a
```

```python
raw_data = {
        'subject_id': ['4', '5', '6', '7', '8'],
        'first_name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'], 
        'last_name': ['Bonder', 'Black', 'Balwner', 'Brice', 'Btisan']}
df_b = pd.DataFrame(raw_data, columns = ['subject_id', 'first_name', 'last_name'])
df_b
```

```python
pd.merge(df_a, df_b, on='subject_id') # subject_id가 있다면 그걸 기준으로 합치기
```

```python
pd.merge(df_a, df_b, left_on='subject_id', right_on='subject_id') # 왼-오른쪽 테이블의 컬럼명이 다르다면 어떤 이름인지 지정해준다
```

```python
pd.merge(df_a, df_b, on='subject_id', how='left') # left를 기준으로 조인함
```

```python
pd.merge(df_a, df_b, on='subject_id', how='right') # right을 기준으로 조인함, inner이 default값임
```

```python
pd.merge(df_a, df_b, on='subject_id', how='outer') # outer join(합집합) 모두다 살린다
```

```python
pd.merge(df_a, df_b, right_index=True, left_index=True) #양옆의 index값을 기준으로 머지해줌
# subject_id_x, y 이렇게 되서 둘 다 삭제해줘야하는 경우도 있음
```

## Concat
- 값을 붙인다

```python
raw_data = {
        'subject_id': ['1', '2', '3', '4', '5'],
        'first_name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'], 
        'last_name': ['Anderson', 'Ackerman', 'Ali', 'Aoni', 'Atiches']}
df_a = pd.DataFrame(raw_data, columns = ['subject_id', 'first_name', 'last_name'])
df_a
```

```python
raw_data = {
        'subject_id': ['4', '5', '6', '7', '8'],
        'first_name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'], 
        'last_name': ['Bonder', 'Black', 'Balwner', 'Brice', 'Btisan']}
df_b = pd.DataFrame(raw_data, columns = ['subject_id', 'first_name', 'last_name'])
df_b
```

```python
df_new = pd.concat([df_a, df_b])
df_new.reset_index()
```

```python
df_new = pd.concat([df_a, df_b])
df_new.reset_index(drop=True) # index 합치기
```

```python
df_a.append(df_b)
```

```python
df_new = pd.concat([df_a, df_b], axis=1)
df_new.reset_index(drop=True)
```
