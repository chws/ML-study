## lambda, map, apply
- pandas의 series type의 데이터에도 map 함수 사용 가능
- function 대신 dict, sequence 형 자료 등으로 대체 가능

```python
s1 = Series(np.arange(10))
s1.head(5)
```

```python
s1.map(lambda x: x**2).head(5)
```

```python
z = {1: 'A', 2: 'B', 3: 'C'} # dict type으로 데이터 교체, 없는 값은 NaN
s1.map(z).head(5)
```

```python
s2 = Series(np.arange(10, 20))
s1.map(s2).head(5) # 같은 위치의 데이터를 s2로 전환
```

```python
df = pd.read_csv("wages.csv")
df.head()
```

```python
df.sex.unique()
```

```python
df["sex_code"] = df.sex.map({'male':0, 'female':1})
df.head(5)
```

```python
df.sex.replace(
    {'male':0, 'female':1}
).head() # dict type 적용
```

```python
df.sex.replace(
    ['male', 'female'], # Target list
    [0, 1], inplace=True) # Conversion list
df.head(5)
```

```python
df_info = df[['earn', 'height', 'age']]
df_info.head()
```

```python
f = lambda x: x.max() - x.min()
df_info.apply(f) # 각 column 별로 결과값 반환
```

```python
df_info.sum()
```

```python
df_info.apply(sum)
```

```python
def f(x):
    return Series([x.min(), x.max()], index=["min", "max"])
df_info.apply(f) # column 별로
```

```python
f = lambda x: -x
df_info.applymap(f).head(5) # 모든 값에 적용
```

```python
f = lambda x: -x
df_info['earn'].apply(f).head(5) # apply를 컬럼 하나에만 적용하면 Map과 거의 같다
```
