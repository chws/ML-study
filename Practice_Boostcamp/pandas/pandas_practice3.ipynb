## Groupby 1

```python
import pandas as pd
```

```python
ipl_data = {
    "Team": [
        "Riders",
        "Riders",
        "Devils",
        "Devils",
        "Kings",
        "kings",
        "Kings",
        "Kings",
        "Riders",
        "Royals",
        "Royals",
        "Riders",
    ],
    "Rank": [1, 2, 2, 3, 3, 4, 1, 1, 2, 4, 1, 2],
    "Year": [2014, 2015, 2014, 2015, 2014, 2015, 2016, 2017, 2016, 2014, 2015, 2017],
    "Points": [876, 789, 863, 673, 741, 812, 756, 788, 694, 701, 804, 690],
}
```

```python
df = pd.DataFrame(ipl_data)
df.groupby("Team")["Points"].mean()
```

```python
df
```

## Hierarchical index
- 두 개의 인덱스

```python
h_index = df.groupby(['Team', 'Year'])['Points'].sum()
h_index
```

```python
h_index.index
```

```python
h_index['Devils':'Kings']
```

```python
h_index.unstack() # data를 matrix형태로 만들어주기도 한다
```

```python
h_index.reset_index()
```

```python
h_index.swaplevel()
```

```python
h_index.swaplevel().sort_index(level=1)
```

```python
type(h_index)
```

```python
h_index.sum(level=0) # level지정만 잘 해주면 ㅇㅋ
```

## grouped

```python
grouped = df.groupby('Team')
```

```python
grouped # generator 형태이므로 데이터가 아직 생성이 안되어있다
```

```python
for name, group in grouped: # 튜플 형태로
    print(name) # 그룹의 key값
    print(group) # value값 추출됨
```

```python
type(group)
```

```python
grouped.get_group('Devils')
```

```python
grouped.agg(sum) # aggregation: 요약된 통계정보를 추출해줌
```

```python
import numpy as np
grouped.agg(np.mean)
```

```python
# Transformation: agg와 달리 key별로 요약된 정보가 아님. 해당 정보를 변환해줌
# Filtration: 특정 정보를 제거하여 보여주는 필터링 기능
grouped['Points'].agg([np.sum, np.mean, np.std]) # 특정 컬럼에 여러개의 function을 apply할 수 있다
```

```python
grouped.agg(max)
```

```python
score = lambda x: (x)
grouped.transform(score) # transform은 각 column별로 모든 값에 동일한 연산이 적용된다
```

```python
df.groupby('Team').filter(lambda x: len(x) >= 3) # 팀별로 세 개 이상의 데이터가 존재하는 것만
```
